This is a notebook that I made to explore a heart disease dataset available here: https://www.kaggle.com/ronitf/heart-disease-uci

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
df = pd.read_csv('heart.csv')
```

```python
correlations = df.corr()
```

```python
#How many heart disease cases do we have in this set?
df['target'].value_counts()
```

```python
#break that out by sex
df.groupby('sex')['target'].value_counts()
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x="sex", hue="target", data=df)
```

I noticed something interesting: the different sexes appear to have a different distribution of target. So next I wanted to look at

```python
sns.boxplot(df['age'])
plt.show()
```

```python
plt.hist(df['age'])
plt.show()
```

```python
sns.distplot(df['age'])
```

```python
#generate KDE plots looking at age distribution by target label

#plt.figure(figsize = (10, 8))

sns.kdeplot(df.loc[df['target'] == 0, 'age'], label = 'Heart Disease')
sns.kdeplot(df.loc[df['target'] == 1, 'age'], label = 'No Heart Disease')

# Labeling of plot
plt.xlabel('Age (years)') 
plt.ylabel('Density') 
plt.title('Distribution of Ages')

plt.show()
```
