### Step 8.  For each occupation present the percentage of women and men

```python
occupation_by_gender = pd.crosstab(users['occupation'],users['gender']).apply(lambda x: x/x.sum()*100, axis=1)
occupation_by_gender
```

```python
res= pd.crosstab(users['occupation'],users['gender'])
#group_gender_data
sums = res.sum(axis=1)
#sums
res['FemaleRatio'] = res['F']*100 / sums
res['MaleRatio'] = res['M'] *100/ sums
res.sort_values('MaleRatio',ascending=False)
```
