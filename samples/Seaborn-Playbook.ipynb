```python
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=200)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', hue='sex', palette='coolwarm', aspect=2)
```
